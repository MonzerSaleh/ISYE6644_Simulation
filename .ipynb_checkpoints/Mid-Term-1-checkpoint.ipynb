### Question 1

**Toss two die and observe their sum. What is the expected number of tosses until the sum is 11??**

If you think about this there is really only 2 events that lead to an 11, (5,6) & (6,5)    
But the question ask how many tosses until this event, so really there's two parts.

**Part 1**    
The probability of a sum of 11 is 2/36, 
- where 2 is the number of successful sums of 11 and 36 is the total number of possibilities.
- this is just 1/18


**Part 2** This is much trickier.    
Consider for example: For a single die, what is the expected number of tosses until we get 4?   
Since P(4)=1/6 we $E[X=x]=1/6$ which leads to 1/1/6 or 6.

So since P(11)=1/18 we compute 1/1/18 which is just 18.

You could have also used a geometric distribution, where E[X]=1/p

### Question 5

Suppose X has PDF $f(x)=4x^3; \; 0 \lt x \lt 1$. Find $E[(2/X)-3]$.   

The traditional path is to integrate but this can be tedious   
$E[(2/X)-3] = \int_0^1 ((2/X)-3)4x^3 dx  $

Another approach is to observe that $E[(2/X)-3] = 2E[1/x]-3$ and this is much easier to integrate    
$E[1/X] = \int_0^1 (1/X)4x^3 dx = \int_0^1 4x^2 dx = (4/3)x^3 |_0^1 = (4/3)$   

So $ 2E[1/x]-3 =  2(4/3)-3 = 8/3 - 9/3 = -1/3$

### Question 6

**Suppose X has PDF $f(x)=3x^2; 0 \le x \le 1$ What's the pdf of the RV $e^X$**

The obvious answer might be $e^{3x^2}; 0 \le x \le 1$

**Approach 1**       
Let $Y = e^x$ we aim to find the cdf of Y and then differentiate it to get the pdf.     
The cdf of Y is   
$
\begin{split}
Y & = G(y)         \\
& = P(Y \le y)     \\
& = P(e^x \le y)   \\
& = P(ln(e^x) \le ln(y) \\
& = P(x \le ln(y)) \\
& = \int 3x^2 & \text{ from 0 to ln(y)}  \\
& = (ln(y))^3  \\
& = \text{ now differentiate to get } \\
& = 3(ln(y))^2 / y  \\
& \text{ from 0 to e } 
\end{split}
$

**Approach 2**    
Let $Y=g(y)$ and note that Y is monotonic   
t/f $F_Y(y) = f_X(g^{-1}(y)) |\frac{d}{dy} (g^{-1}(y)) |$

So we compute each part   
- $Y = e^x$; we solve for x to get $ln(y)=ln(e^x)=x$    
- $x = g^{-1}(y) = ln(y)$    
    - derivative of x is given by $\frac{d}{dy} ln(y) = \frac{1}{y}$

$F_Y(y) = 3(ln(y))^2 \cdot | \frac{1}{y} | $ which is the same as previous method


### Question 7

**Suppose that X and Y have a joint pdf f(x,y)=8xy, 0< y < x < 1**
- Find E[X]
- Find Cov(X,Y)

Given the joint pdf we need to determine the marginal for each    
$f_X(x) = \int_0^x f(x,y) dy = \int_0^x 8xy dy = 4xy^2 |_0^x = 4x^3$     
Similary $f_Y(y) = \int_y^1 f(x,y) dx = \int_y^1 8xy dx = 4x^2y |_y^1 = 4y - 4y^3$     

Now we compute   
$E[X] = \int_0^1 4x^3 x dx = 4x^5 / 5 |_0^1 = 4/5$    
Similarly $E[XY] = 8/15$
But also   
$E[XY]=\int_0^1 \int_0^x (xy)(8xy) dy dx = \cdots \text{ details left out} = 8/18 = 4/9 $


**Covariance**   
Recall that covariance can be computed in 2 ways   
1. Cov(XY) = E[XY] - E[x] E[y]
2. Cov(XY) = E[(X-E[X])(Y-E[Y])]

Using the first    
$Cov(XY) = E[XY] - E[x] E[y] = (4/9) - (4/5)(8/15) = 4/225$

### Question 8

**Let X,Y be iid exponential with rate 1. Find P(X+Y >= 2)**    

Consider $Z=X+Y$ then $Z=Exp(\lambda=1)+Exp(\lambda=1)$ this has an Erlang distribution $Erlang(k=2,\lambda=1)$         
So we can reframe the problem as find $P(Erlang(2,1) ge 2) \; or \; 1-F(2)$ where F(x) is the cdf        
Recall that the cdf of an erlang is just $F(x,k,\lambda)=1-\sum_0^{k-1} \frac{1}{i!} e^{-\lambda x} (\lambda x)^i$     

Now we can compute with k=2,$\lambda=1$   
$= 1-(1-\sum_0^{1} \frac{1}{1!} e^{-1x} (1x)^i)$      
$= \sum_0^{1} e^{-x}(x)^i$      
$= e^{-x}(x)^0+ e^{-x}(x)^1$    
$= e^{-x}( 1 + x )$     
$= e^{-2}( 1 + 2 )$     
$= 3e^{-2} = 0.406 $     


### Question 9

**Suppose X is a normal RV with mean 1 and variance 9. What us the probability that $X \ge 4$**    

This is a pretty standard type of question. we're given that $X \sim N(1,9)$        
We also note that the variance 9 has a standard deviation of 3         
So what is $P(N(1,9) \ge 4)$ ?

This is where we need to compute the Z score for a standard normal       
$Z=\frac{x-\mu}{\sigma}=\frac{x-1}{3}$          
So $P(N(1,9) \ge 4) = Pr(\frac{x-1}{3} \ge (4-1)/3 = 1 ) = Pr(Z \ge 1)$             
Which is $1-Pr(Z \le 1)$              
Which we can look up in a normal table as $1-0.8413=0.1587$



### Question 10

**If X,Y are iid Nor(0,1) RVs, find P(Y < 2X + 1)**    

First we move the X to the left side to get $P(Y - 2X < 1)$           
Let   
$
\begin{split}
Z & = Y - 2X                        \\
& = N(0,1) - 2*N(0,1)               \\
& = N(0,1) - N(0,4)                 \\
& = N(0,5)  & \text{ because iid }
\end{split}
$

So now the question is : What is $Pr(N(0,5) < 1)$? Well similar to the previous problem                 
$P(\frac{N(0,5)-0}{\sqrt{5}} \le \frac{1-0}{\sqrt{5}})$          
$=P(Z \le \frac{1}{\sqrt{5}})$          
$=P(Z \le 0.447)$          
$\approx 0.67$

### Question 13   

**Suppose that the Atlanta Hawks play iid games, each of which has win probability of 0.6. Let X be the number of games until the hawks achieve their first win. Find the smallest x such that $P(X \le x) \ge 0.9$**

My Approach    
The problem is formulated as a geometric one $G(p;x)=(1-p)^x p$ with expected value 1/p and variance 1-p/p^2          
The cdf is given by $1-(1-p)^x$             

So $P(X \le x) \ge 0.9 \implies F(x) \ge 0.9 \implies 1-(1-p)^x \implies 1-(1-0.6)^x$            
Now it's just matter of trial and error        
F(X=0)=0     > .9 No         
F(X=1)=0.6   > .9 No           
F(X=2)=0.84  > .9 No           
F(X=3)=0.936 > .9 YES we have a winner          



### Question 14     

**Suppose X1,...,Xn are iid from some distribution. What tells us that the sample mean of the Xi(s) is approximately normal for large enough n?**

The mean, of Z where $X=X_1+\cdots+X_n$ is defined as $\mu_Z = \sum_1^n X_i = n \mu_X $         
Similarly the variance is $n\sigma^2$             
This $Z \sim N(n\mu, n\sigma^2)$
